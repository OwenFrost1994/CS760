```python
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
```

```python
tmp = []
with open('HW3/D2a.txt', 'r') as file:
    for row in file:
        row = row.replace('\n', '')
        tmp.append(row.split(' '))
D2a = np.array(tmp).astype('float')
D2a
```

```python
D2aF = pd.DataFrame(data=D2a, columns=["x1", "x2", "x3", "x4","x5", "x6", 'y'])
D2aF
```

```python
tmp = []
with open('HW3/D2z.txt', 'r') as file:
    for row in file:
        row = row.replace('\n', '')
        tmp.append(row.split(' '))
D2z = np.array(tmp).astype('float')
D2z
```

```python
D2zF = pd.DataFrame(data=D2z, columns=["x1", "x2", 'y'])
D2zF
```

```python
np.mean(D2zF.iloc[:,0:2])
```

```python
np.var(D2zF.iloc[:,0:2])
```

```python
x = np.arange(-2, 2.1, 0.1)
y = np.arange(-2, 2.1, 0.1)
x, y
```

```python
tmp = []
for i in range(x.shape[0]):
    for j in range(y.shape[0]):
        tmp.append([x[i],y[j]])
D2t = np.array(tmp).astype('float')
D2t
```

```python
fig,ax = plt.subplots(1,1,figsize=(8,8))
ax.scatter(D2zF.loc[D2zF.iloc[:,2] == 1].iloc[:,0], D2zF.loc[D2zF.iloc[:,2] == 1].iloc[:,1], c = 'r', s=10, alpha=0.5)
ax.scatter(D2zF.loc[D2zF.iloc[:,2] == 0].iloc[:,0], D2zF.loc[D2zF.iloc[:,2] == 0].iloc[:,1], c = 'b', s=10, alpha=0.5)
ax.scatter(D2t[:,0],D2t[:,1], c = 'k', s=10, alpha=0.5)
ax.set_xlabel('$x_1$', fontsize=20)
ax.set_ylabel('$x_2$', fontsize=20)
ax.set_xlim([-2,2])
ax.set_ylim([-2,2])
```

```python
D2zF['d'] = 0
Predic_label = []
for i in range(D2t.shape[0]):
    diff = D2t[i,:]-D2zF.iloc[:,0:2]
    for j in range(D2zF.shape[0]):
        D2zF.iloc[j, 3] = np.array(diff.iloc[j,:]).dot(np.diag(np.var(D2zF.iloc[:,0:2]))).dot(diff.iloc[j,:].T)
    
    D2zF_sort = D2zF.sort_values(by='d', axis = 0, ascending = True)
    Predic_label.append(D2zF_sort.iloc[0:3,2].value_counts().index[0])
```

```python
D2tPDF = pd.DataFrame(data=D2t, columns=["x1", "x2"])
D2tPDF['y'] = Predic_label
D2tPDF
```

```python
fig,ax = plt.subplots(1,1,figsize=(8,8))
ax.scatter(D2zF.loc[D2zF.iloc[:,2] == 1].iloc[:,0], D2zF.loc[D2zF.iloc[:,2] == 1].iloc[:,1], c = 'r', s=10, alpha=1)
ax.scatter(D2zF.loc[D2zF.iloc[:,2] == 0].iloc[:,0], D2zF.loc[D2zF.iloc[:,2] == 0].iloc[:,1], c = 'b', s=10, alpha=1)
ax.scatter(D2tPDF.loc[D2tPDF.iloc[:,2] == 1].iloc[:,0], D2tPDF.loc[D2tPDF.iloc[:,2] == 1].iloc[:,1], c = 'r', s=10, alpha=0.2)
ax.scatter(D2tPDF.loc[D2tPDF.iloc[:,2] == 0].iloc[:,0], D2tPDF.loc[D2tPDF.iloc[:,2] == 0].iloc[:,1], c = 'b', s=10, alpha=0.2)
ax.set_xlabel('$x_1$', fontsize=20)
ax.set_ylabel('$x_2$', fontsize=20)
ax.set_xlim([-2,2])
ax.set_ylim([-2,2])
ax.legend(['1 training','0 training','1 testing','0 testing'], fontsize=20)
```

```python
D2aFR = D2aF.sample(frac=1)
```

```python
FeatDim = 6
Errn = []
for n in range(4):
    test = D2aFR.iloc[50*n:50*(n+1),:]
    train = D2aFR.drop(D2aFR.index[range(50*n,50*(n+1))])
    
    train['d'] = 0
    Predic_label = []
    for i in range(test.shape[0]):
        diff = test.iloc[i,0:FeatDim]-train.iloc[:,0:FeatDim]
        for j in range(diff.shape[0]):
            train.iloc[j, -1] = np.array(diff.iloc[j,:]).dot(diff.iloc[j,:].T) ** 0.5
        
        train_sort = train.sort_values(by='d', axis = 0, ascending = True)
        Predic_label.append(train_sort.iloc[0:3,-2].value_counts().index[0])
        
    Errn.append(sum(test['y'] != Predic_label))
    
```

```python
Errn
```

```python
Errr = []
for i in range(len(Errn)):
    Errr.append(100 * Errn[i] /50) 

print("Error rate for each fold:", Errr)
print("Average error rate is", sum(Errr)/4)
```
