```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sample_n = 100
mu = np.array([0, 0])
Sigma = np.array([[1,0],[0,1]])
X = np.random.multivariate_normal(mu,Sigma,sample_n)
```

```python
fig,ax = plt.subplots(1,1,figsize=(8,8))
ax.scatter(X[:,0], X[:,1], s=10, alpha=0.5)
ax.set_xlabel('$x_1$', fontsize=20)
ax.set_ylabel('$x_2$', fontsize=20)
ax.set_xlim([-3,3])
ax.set_ylim([-3,3])
```

```python
X1 = np.random.multivariate_normal(mu + [1,-1],2*Sigma,sample_n)
```

```python
fig,ax = plt.subplots(1,1,figsize=(8,8))
ax.scatter(X1[:,0], X1[:,1], s=10, alpha=0.5)
ax.set_xlabel('$x_1$', fontsize=20)
ax.set_ylabel('$x_2$', fontsize=20)
ax.set_xlim([-3,3])
ax.set_ylim([-3,3])
```

```python
X2 = 0.3*np.random.multivariate_normal(mu + [1,0],Sigma+[[0,0.2],[0.2,0]],sample_n) \
    + 0.7*np.random.multivariate_normal(mu + [-1,0],Sigma+[[0,-0.2],[-0.2,0]],sample_n)
```

```python
fig,ax = plt.subplots(1,1,figsize=(8,8))
ax.scatter(X2[:,0], X2[:,1], s=10, alpha=0.5)
ax.set_xlabel('$x_1$', fontsize=20)
ax.set_ylabel('$x_2$', fontsize=20)
ax.set_xlim([-3,3])
ax.set_ylim([-3,3])
```
