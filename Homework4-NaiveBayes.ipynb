```python
from glob import glob
import pandas as pd
import math
```

```python
dirs = glob('HW4/*')
eng = glob('HW4/e/train/*.txt')
japa = glob('HW4/j/train/*.txt')
span = glob('HW4/s/train/*.txt')
```

```python
alpha = 0.5
KL = 3
```

```python
phat_e = (len(eng) + alpha) / (len(eng) + len(japa) + len(span) + KL * alpha)
phat_e
```

```python
phat_j = (len(japa) + alpha) / (len(eng) + len(japa) + len(span) + KL * alpha)
phat_j
```

```python
phat_s = (len(span) + alpha) / (len(eng) + len(japa) + len(span) + KL * alpha)
phat_s
```

```python
results_e = []
for i in range(len(eng)):
    tmp = ''
    with open(eng[i], 'r') as file:
        for row in file:
            row = row.replace('\n', '')
            tmp += row
    
    result={}
    for j in tmp:
        result[j] = tmp.count(j)
    result['space'] = result.pop(' ')
    
    results_e.append(result)
```

```python
from random import shuffle
```

```python
def shuffle_str(s):
    str_list = list(s)
    shuffle(str_list)
    return ''.join(str_list)
```
